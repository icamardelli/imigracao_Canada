Uma empresa especializada em consultoria e aconselhamento para pessoas que desejam migrar do Brasil para o Canadá está buscando melhorar seus serviços. Para isso, foram disponibilizados alguns arquivos para analisar as tendências de imigração do Brasil nos últimos anos.

##Explorando e tratando os dados

Importando a biblioteca Pandas para manipulação e tratamento de dados:

```python
import pandas as pd
```

Armazenando o link do arquivo .CSV na variável URL:

```python
url = 'https://raw.githubusercontent.com/alura-cursos/bibliotecas_visualizacao/refs/heads/main/Dados/imigrantes_canada.csv'
```

Lendo o arquivo .CSV passando como parâmetro a URL:

```python
df = pd.read_csv(url)
df
```

Para obter algumas informações mais detalhadas, utilizou-se o método `.info( )`:

```python
df.info()
```

##Análisando as tendências de imigração do Brasil ao longo dos anos:

Setando o índice como a coluna "País" e passando o parâmetro inplace=True para que a alteração seja aplicada diretamente ao DataFrame, sem necessidade de atribuições:

```python
df.set_index('País', inplace=True)
```

Armazenando uma lista com todos os anos contidos no DataFrame:

```python
anos = list(map(str, range(1980, 2014)))
```

Verificando a variável criada:

```python
anos
```

Selecionando apenas os dados do Brasil e passando como segundo parâmetro (colunas) a variável criada acima contendo apenas os anos:

```python
brasil = df.loc['Brasil', anos]
```

Verificando o intervalo selecionado:

```python
brasil
```

Criando um DataFrame com os dados de anos e quantidade de imigrantes via dicionário:

Obs.: **Aplicação do método `.tolist( )` para converter um objeto Series em uma lista.**

```python
dados_brasil_dict = {
    'ano': brasil.index.tolist(),
    'imigrantes': brasil.values.tolist()
}

dados_brasil = pd.DataFrame(dados_brasil_dict)
```

```python
dados_brasil
```

###Ranqueando países da América do Sul

Veriicando o DataFrame:

```python
df.head()
```

Aplicando o método `.query( )` para filtrar a coluna Região como América do Sul:

```python
america_sul = df.query('Região == "América do Sul"')
america_sul
```

Plotando via gráfico de barras horizontais o comparativo entre os países da América do Sul:

```python
cores = ['royalblue', 'orange', 'forestgreen', 'orchid', 'purple', 'brown', 'slateblue', 'gray', 'olive', 'navy', 'teal', 'tomato']

fig, ax = plt.subplots(figsize=(12,5))

ax.barh(america_sul.index, america_sul['Total'], color=cores)
ax.set_title('Migração dos países da América do Sul para o Canadá\n1980 a 2013', loc='left', fontsize=18)
ax.set_xlabel('Qtd. de Imigrantes', fontsize=14)
ax.set_ylabel('')
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

plt.show()
```

Organizando o DataFrame através do método `.sort_values( )`:

```python
america_sul_sorted = america_sul.sort_values('Total')
```

Plotando novamente o comparativo entre países da América do Sul, destacando o Brasil através do loop `for` e o condicional `if`.

Aplicando também a rotulação das barras através da função `enumerate( )`:

```python
cores = []
for pais in america_sul_sorted.index:
  if pais == 'Brasil':
    cores.append('green')

  else:
    cores.append('silver')

fig, ax = plt.subplots(figsize=(12,5))
ax.barh(america_sul_sorted.index, america_sul_sorted['Total'], color=cores)
ax.set_title('América do Sul: Brasil foi o quarto país com mais imigrantes \npara o Canadá entre 1980 e 2013', loc='left', fontsize=18)
ax.set_xlabel('Qtd. de Imigrantes', fontsize=14)
ax.set_ylabel('')
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

for i, v in enumerate(america_sul_sorted['Total']):
  ax.text(v + 500, i, str(v), color='black', fontsize=10, ha='left', va='center')

ax.set_frame_on(False)
ax.get_xaxis().set_visible(False)
ax.tick_params(axis='both', which='both', length=0)

fig.savefig('imigracao_america_sul_canada.png', transparent=False, dpi=300, bbox_inches='tight')

plt.show()
```

###Visualizando TOP 10 países com mais imigrantes para o Canadá

Criando um novo DataFrame atraves de uma cópia do DataFrame (método `.copy( )`) original, organizando os dados de forma decrescente através do método `.sort_values( )` e parâmetro `ascending=False` e capturando os 10 primeiros países através do método `.head( )` passando o 10 como parâmetro:

```python
top_10 = df.sort_values('Total', ascending=False).head(10).copy()
top_10
```

Criando um gráfico de barras horizontais para melhor visualização dos dados:


```python
sns.barplot(data=top_10, x='Total', y=top_10.index)
```

Adicionando elementos (rótulos, títulos etc.):

```python
ax = sns.barplot(data=top_10, x='Total', y=top_10.index)

ax.set(title='Países com maior imigração para o Canadá\n1980 a 2013',
       xlabel='Qtd. de Imigrantes',
       ylabel='')

plt.show()
```

Criando figuras e mixando as biblioteas Seaborn com Matplotlib:

```python
fig, ax = plt.subplots(figsize=(10,6))
ax = sns.barplot(data=top_10, x='Total', y=top_10.index)

ax.set_title('Países com maior imigração para o Canadá\n1980 a 2013', loc='left', fontsize=16)
ax.set_xlabel('Qtd. Imigrantes', fontsize=14)
ax.set_ylabel('')

plt.show()
```

Escolhendo a paleta de cores mais adequada aos dados através da criação de uma função:

`hue=top_10.index`: Indica que as barras serão agrupadas e coloridas de acordo com os valores dos índices do DataFrame top_10. Cada grupo terá uma cor distinta. Necessário passar o `hue` juntamente ao `palette`.

`palette='Set2'`: Aqui, é especificado a paleta de cores a ser usada para colorir as barras.

_________________________

Documentação Seaborn para paleta de cores: https://seaborn.pydata.org/tutorial/color_palettes.html

```python
def gerar_grafico_paleta(palette):
  fix, ax = plt.subplots(figsize=(8,4))
  ax = sns.barplot(data=top_10, x='Total', y=top_10.index, hue=top_10.index, palette=palette)
  ax.set_title('Países com maior imigração para o Canadá\n1980 a 2013', loc='left', fontsize=16)
  ax.set_xlabel('Qtd. Imigrantes', fontsize=14)
  ax.set_ylabel('')

  plt.show()
```

Testando as paletas sequenciais:

```python
gerar_grafico_paleta('Blues_r')
```

Testando as paletas categóricas:

```python
gerar_grafico_paleta('tab10')
```

Explorando temas:

```python
sns.set_theme(style='ticks')
gerar_grafico_paleta('tab10')
```

Removendo as bordas superiores e lateral direita utilizando a função `despine( )`:

```python
fig, ax = plt.subplots(figsize=(8,4))
ax = sns.barplot(data=top_10, x='Total', y=top_10.index, hue=top_10.index, palette='tab10')
ax.set_title('Países com maior imigração para o Canadá\n1980 a 2013', loc='left', fontsize=16)
ax.set_xlabel('Qtd. Imigrantes', fontsize=14)
ax.set_ylabel('')
sns.despine()

plt.show()
```

##Data Visualization - Plotly

Importando a biblioteca Plotly:

```python
import plotly.express as px
```

Plotando um gráfico de linhas interativo, utilizando o DataFrame com os dados do Brasil:

```python
fig = px.line(dados_brasil, x='ano', y='imigrantes',
              title='Imigração do Brasil para o Canadá de 1980 a 2013')
fig.update_layout(width=1000,
                  height=500,
                  xaxis={'tickangle':-45},
                  font_family='Arial',
                  font_size=14,
                  font_color='gray',
                  title_font_color='black',
                  title_font_size=25,
                  xaxis_title='Ano',
                  yaxis_title='Qtd. Imigrantes')
fig.show()
```

Atribuindo cores ao gráfico:

```python
fig = px.line(dados_brasil, x='ano', y='imigrantes',
              title='Imigração do Brasil para o Canadá de 1980 a 2013')

fig.update_traces(line_color='green', line_width=4)

fig.update_layout(width=1000,
                  height=500,
                  xaxis={'tickangle':-45},
                  font_family='Arial',
                  font_size=14,
                  font_color='gray',
                  title_font_color='black',
                  title_font_size=25,
                  xaxis_title='Ano',
                  yaxis_title='Qtd. Imigrantes')
fig.show()
```
